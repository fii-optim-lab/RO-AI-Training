## Containers

### Lists

The simplest container is a Python `list`.

```python
vec = [1, 2, 3]
vec
```

```python
empty = []
empty
```

Lists can have elements of different type

```python
a_list = [1, 2, 2, 3.0, '5']
print(a_list)
print(type(a_list))
```

```python
a_list[0]
```

```python
a_list[0] = 0
print(a_list)
```

### Tuples

Tuples are similar to lists, but the elements are immutable.

```python
a_tuple = tuple(a_list)
print(a_tuple)
print(type(a_tuple))
```

### Set

Sets have unique elements

```python
a_set = {1, 2, 3, 4}
print(a_set)
```

```python
a_set = set(a_list)
print(a_set)
print(type(a_set))
```

### Dicts

Dictionaries are maps with unique keys. Each key has a corresponding value.

```python
a_dict = {
    1: 'unu',
    'doi': 2,
    3.0: a_set
}
a_dict
```

```python
print("Keys:\t", list(a_dict.keys()))
print("Values:\t", list(a_dict.values()))
print(type(a_dict))
```

## Functions

```python
def add_two_values(a, b):
    return a + b

add_two_values(5, 6)
```

```python
def multiply(a: float, b: float) -> float:
    return a * b

multiply(4.0, 5.0)
```

```python
multiply("AB", 2)
```

## Classes

```python
class Multiplier:
    def __init__(self, x):
        self.x = x

    def multiply(self, y):
        return self.x * y
```

```python
m = Multiplier(5)
print(m.multiply(10))
print(m.multiply(3))
print(m.multiply(0))
```

## Files

```python
with open("file.txt", "w") as f:
    f.write("This is a line\nOn a file\n")
    f.write("This is another\n\tline   with spaces")
```

```python
with open("file.txt", "r") as f:
    print(f.read())
```

```python
with open("file.txt", "r") as f:
    lines = f.readlines()
print(lines)
for line in lines:
    print(line)
```

```python
for line in lines:
    print(line.strip("\n"))
```
